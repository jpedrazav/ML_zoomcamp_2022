### Question 7
Select all the "Lotus" cars from the dataset.

Select only columns "Engine HP", "Engine Cylinders".

Now drop all duplicated rows using drop_duplicates method (you should get a dataframe with 9 rows).

Get the underlying NumPy array. Let's call it X.

Compute matrix-matrix multiplication between the transpose of X and X. To get the transpose, use X.T. Let's call the result XTX.

Invert XTX.

Create an array y with values [1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800].

Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.

What's the value of the first element of w?

```python
# https://re-thought.com/pandas-value_counts/
# PROC Freq by 2 variables

data_lotus_test=data[data['Make']=='Lotus']
data_lotus_test.groupby('Engine HP')['Engine Cylinders'].value_counts()
```

```python
data_lotus = data[data['Make'] == 'Lotus'][['Engine HP','Engine Cylinders']].drop_duplicates()
data_lotus
```

```python
#convert to numpy array
X = data_lotus.to_numpy()
X
```

```python
#transpose of X
XT = X.T
XT
```

```python
#matrix multiplication
XTX = np.dot(XT, X)
XTX
```

```python
#Invert of the transpose of XTX
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
#new array y
y = [1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800]
```

```python
W = np.dot(XTX_inv,XT).dot(y)
W[0]
```
